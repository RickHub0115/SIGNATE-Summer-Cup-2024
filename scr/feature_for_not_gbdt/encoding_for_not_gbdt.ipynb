```python
# すべてのカテゴリ変数から、2つ組み合わせた特徴量
print(len(df_train.columns[44:110]))
print(df_train.columns[44:110])

combination_columns = df_train.columns[44:109]
```

```python
# カテゴリ変数

columns = [
    'TypeofContact',
    'CityTier',
    'Occupation',
    'Gender',
    'ProductPitched',
    'PreferredPropertyStar',
    'Passport',
    'PitchSatisfactionScore',
    'Designation',
    'Marry',
    'Car',
    'Child',
    # 以下作成特徴量
    'AgeGroup',
    'TypeofContactNULL',
    'Child01',
    #'EconomicSegment', # target-encodingの方がGood
    'IsFamily',
    'FreaqencySeg',
    'MonetarySeg',
    ]
```

```python
df_train_encoded = pd.get_dummies(df_train, columns=columns, drop_first=True, dtype=int)
df_test_encoded = pd.get_dummies(df_test, columns=columns, drop_first=True, dtype=int)

df_train_encoded = pd.get_dummies(df_train_encoded, columns=combination_columns, drop_first=True, dtype=int)
df_test_encoded = pd.get_dummies(df_test_encoded, columns=combination_columns, drop_first=True, dtype=int)
```

```python
# 特徴量の調整
df_train_aligned, df_test_aligned = df_train_encoded.align(df_test_encoded, join='inner', axis=1)
```

```python
print(len(df_train_aligned.columns))
print(len(df_test_aligned.columns))
```
