```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scr.util import *
```

```python
df_train = pd.read_csv('data/feature_engineered/train_feature_ok.csv')
df_test = pd.read_csv('data/feature_engineered/test_feature_ok.csv')
```

```python
df_train['ProdTaken'].value_counts()
```

```python
def feature_visualization_str(feature_str):
    feature = feature_str

    tmp_0 = df_train.loc[df_train['ProdTaken'] == 0, feature]
    tmp_1 = df_train.loc[df_train['ProdTaken'] == 1, feature]
    tmp_list = [tmp_0, tmp_1]

    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # カテゴリの順序を統一するために、全データのユニークな値を取得
    unique_categories = sorted(df_train[feature].dropna().unique())

    for i, tmp in enumerate(tmp_list):
        # カテゴリの順序を統一する
        tmp = pd.Categorical(tmp, categories=unique_categories, ordered=True)
        sns.histplot(x=tmp, ax=axes[i])
        axes[i].set_title(f'ProdTaken = {i}')

    plt.show()
```

```python
def feature_visualization(feature_str):
    feature = feature_str

    tmp_0 = df_train.loc[df_train['ProdTaken'] == 0, feature]
    tmp_1 = df_train.loc[df_train['ProdTaken'] == 1, feature]
    tmp_list = [tmp_0, tmp_1]

    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # 全データの範囲を取得し、各ヒストグラムで同じ範囲とビンを使用する
    min_value = df_train[feature].min()
    max_value = df_train[feature].max()
    bins = 30  # ビンの数は任意ですが、全てのヒストグラムで同じにします

    for i, tmp in enumerate(tmp_list):
        sns.histplot(x=tmp, ax=axes[i], bins=bins, binrange=(min_value, max_value))
        axes[i].set_title(f'ProdTaken = {i}')

    plt.show()
```

```python
def feature_statics(feature_str):
    display(df_train.groupby(by=[feature_str])['ProdTaken'].agg(['mean', 'count']).sort_values(by='mean', ascending=False))
```
